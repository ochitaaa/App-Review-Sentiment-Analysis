#**Scraping Dataset**

```python
from google_play_scraper import app, reviews_all, Sort

# Mengambil semua ulasan dari aplikasi dengan ID 'com.mobile.legends' di Google Play Store.
result, continuation_token = reviews(
    'com.mobile.legends',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=12000
)

print(f" Scraping berhasil! Jumlah ulasan yang diambil: {len(result)}")
```

```python
# Lihat 5 review pertama lengkap (dictionary)
for r in result[:5]:
    print(r)

# Atau hanya content + score
print("\n5 Review dengan Score:")
for r in result[:5]:
    print(f"‚≠ê {r['score']} | {r['content']}")
```

```python
# Lihat jumlah total ulasan
print("Jumlah ulasan:", len(result))

# Lihat 1 ulasan pertama (format dictionary lengkap)
print("\nContoh 1 ulasan pertama (full data):")
print(result[0])

# Lihat hanya 5 konten ulasan pertama
print("\n5 Ulasan pertama:")
for review in result[:5]:
    print("-", review['content'])
```

```python
# Ubah hasil scraping (list of dict) jadi DataFrame
df = pd.DataFrame(result)

# Ambil kolom penting (content + score)
df = df[['content', 'score']]

# Simpan ke CSV
df.to_csv("ulasan_aplikasi.csv", index=False, encoding="utf-8")

print(f"Data berhasil disimpan ke 'ulasan_aplikasi.csv' dengan {len(df)} baris.")
```
